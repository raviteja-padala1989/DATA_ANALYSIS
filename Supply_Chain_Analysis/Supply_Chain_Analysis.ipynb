## Top-Performing Products(by Revenue generation)

```python
# Calculate revenue generated per product
df['Revenue per Product'] = df['Revenue generated'] / df['Number of products sold']

# Calculate the cost per product
df['Cost per Product'] = df['Costs'] / df['Number of products sold']

# Calculate profit per product
df['Profit per Product'] = df['Revenue per Product'] - df['Cost per Product']

# Sort products by revenue generated in descending order
top_selling_products = df.sort_values(by='Revenue generated', ascending=False)

# Print the top-performing products
print("Top-Performing Products:")
top_selling_products[['Product type', 'SKU', 'Revenue generated', 'Number of products sold', 'Revenue per Product', 'Cost per Product', 'Profit per Product']].head()
```

*Observation*:

* The top-performing products are all haircare products, with SKU51 being the most profitable. This suggests that haircare products are in high demand and are generating a lot of revenue for the company.

## Most_profitable_products

```python
# Sort products by profit per product in descending order
most_profitable_products = df.sort_values(by='Profit per Product', ascending=False)

# Print the most profitable products
print("\nMost Profitable Products:")
most_profitable_products[['Product type', 'SKU', 'Revenue generated', 'Number of products sold', 'Revenue per Product', 'Cost per Product', 'Profit per Product']].head()
```

*Observation*:

The most profitable products are all haircare products, with SKU2 being the most profitable. This suggests that haircare products are in high demand and are generating a lot of profit for the company.

## Profitability by Product type

```python
# Calculate profitability
df['Profitability'] = df['Revenue generated'] - df['Manufacturing costs']

# Calculate profitability as a percentage
df['Profitability(%)'] = (df['Revenue generated'] - df['Manufacturing costs']) / df['Revenue generated'] * 100

# Group by 'Product type' and calculate the average profitability and profitability percentage for each product type
profitability_by_product_type = df.groupby('Product type').agg({
    'Profitability': 'mean',
    'Profitability(%)': 'mean'
}).reset_index()

# Sort the result by profitability in descending order
profitability_by_product_type = profitability_by_product_type.sort_values(by='Profitability', ascending=False)

# Print the profitability by product type
print(profitability_by_product_type)
```

# Product Availability Percentage

```python

# Step 1: Calculate the availability percentage for each product type
availability_percentage = df.groupby('Product type')['Availability'].mean() * 100

# Step 2: Define a threshold for low availability (e.g., 50%)
low_availability_threshold = 50

# Identify products with low availability
low_availability_products = availability_percentage[availability_percentage < low_availability_threshold]

# Step 3: Take appropriate actions to address low availability issues (e.g., alert or restock)
if not low_availability_products.empty:
    print("Products with Low Availability:")
    print(low_availability_products)


plt.figure(figsize=(12, 4))
availability_percentage.plot(kind='bar', color='skyblue')
plt.title('Product Availability Percentage by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Availability Percentage (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print(f"{availability_percentage}")
```

```python

# Calculate the mean stock level for all products
mean_stock_level = df['Stock levels'].mean()

# Define a threshold for overstocked and understocked products (you can adjust these thresholds)
overstock_threshold = mean_stock_level * 1.2  # Example: 120% of the mean
understock_threshold = mean_stock_level * 0.8  # Example: 80% of the mean

# Create a new column 'Stock Status' to categorize products as 'Overstocked' or 'Understocked'
df['Stock Status'] = 'Normal'  # Initialize all products as 'Normal'

# Mark products as 'Overstocked' or 'Understocked'
df.loc[df['Stock levels'] > overstock_threshold, 'Stock Status'] = 'Overstocked'
df.loc[df['Stock levels'] < understock_threshold, 'Stock Status'] = 'Understocked'

# Group by 'Product type' and count the number of products in each category
stock_status_counts = df.groupby(['Product type', 'Stock Status']).size().unstack(fill_value=0)

# Print the results
print(stock_status_counts)
```
